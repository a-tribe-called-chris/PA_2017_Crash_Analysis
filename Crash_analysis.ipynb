```python
  import pandas as pd
  import matplotlib.pyplot as plt
  pd.set_option('display.max_rows', 100)

def read_csv_file(filename):
    df = pd.read_csv(filename, engine = 'python')
    df.head()
    return df

def plot_histogram(df):
    plt.figure(figsize = (20,10), dpi = 100)
    plt.hist(df['MODEL_YR'], bins = range(1980,2020,1), color = 'blue', edgecolor = 'black', linewidth = 1.2)
    plt.title('Histogram of Crash Frequency by Vehicle Model Year')
    plt.xlabel('Model Year')
    plt.ylabel('Crash Frequency')
    plt.show()

def num_crashes_per_county(df):
    num_crashes_per_county = df.groupby('COUNTY_NAME').size()
    plt.figure(figsize = (20,10), dpi = 100)
    plt.bar(num_crashes_per_county.index, num_crashes_per_county.values)
    plt.title('Number of Crashes per County')
    plt.xlabel('County')
    plt.ylabel('Number of Crashes')
    plt.xticks(rotation = 90)
    plt.show()

def num_crashes_per_county_by_crash_type(df):
    num_crashes_per_county_by_crash_type = df.groupby(['IMPACT_POINT']).size()
    plt.figure(figsize = (20,10), dpi = 100)
    plt.bar(num_crashes_per_county_by_crash_type.index, num_crashes_per_county_by_crash_type.values)
    plt.title('Number of Crashes per County by Crash Type')
    plt.xlabel('County')
    plt.ylabel('Number of Crashes')
    plt.xticks(rotation = 90)
    plt.show()

def num_crashes_by_color(df):
    num_crashes_by_color = df.groupby(['VEH_COLOR_CD']).size()
    plt.figure(figsize = (20,10), dpi = 100)
    plt.bar(num_crashes_by_color.index, num_crashes_by_color.values)
    plt.title('Number of Crashes by Color')
    plt.xlabel('Vehicle Color')
    plt.ylabel('Number of Crashes')
    plt.xticks(rotation = 90)
    plt.show()

def num_crashes_by_CO_miles_drive(df):
    plt.figure(figsize = (20,10), dpi = 100)
    county_name = df.groupby(['COUNTY_NM']).size()
    num_crash_rate = df.groupby(['Rate']).size()
    plt.bar(county_name.index, num_crash_rate.values)
    plt.title('Number of Crashes by County')
    plt.xlabel('County')
    plt.ylabel('Rate of Crashes')
    plt.xticks(rotation = 90)
    plt.show()

def correlation_Body_type_and_impact_point(df):
    corr_factor = df.corr(method = 'pearson')
    return corr_factor

read_csv_file('data_veh.csv')
plot_histogram(read_csv_file('data_veh.csv'))
num_crashes_per_county(read_csv_file('data_veh.csv'))
num_crashes_per_county_by_crash_type(read_csv_file('data_veh.csv'))
num_crashes_by_color(read_csv_file('data_veh.csv'))
num_crashes_by_CO_miles_drive(read_csv_file('data_veh.csv'))
correlation_Body_type_and_impact_point(read_csv_file('data_veh.csv'))
```
